```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create sample data with NumPy
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a DataFrame with Pandas
df = pd.DataFrame({
    'x': x,
    'sin(x)': y,
    'cos(x)': np.cos(x)
})

# Print first few rows
print("DataFrame head:")
print(df.head())

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(df['x'], df['sin(x)'], label='sin(x)')
plt.plot(df['x'], df['cos(x)'], label='cos(x)')
plt.title('Sine and Cosine Waves')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Basic calculations
print("\nNumPy array operations:")
array = np.array([1, 2, 3, 4, 5])
print(f"Mean: {np.mean(array)}")
print(f"Sum: {np.sum(array)}")
print(f"Standard deviation: {np.std(array)}")

# Pandas operations
print("\nPandas DataFrame operations:")
print(f"DataFrame shape: {df.shape}")
print("\nDataFrame description:")
print(df.describe())
```
